```python
def loan_payment(principal, annual_rate, months):
  assert isinstance(principal, (int, float))
  assert isinstance(annual_rate, (int, float))
  assert isinstance(months, int)

  if principal <= 0 or annual_rate < 0 or months <= 0:
      raise ValueError("All inputs must be positive")

  monthly_rate = annual_rate / 100 / 12

  if monthly_rate != 0:
      monthly_payment = (monthly_rate * principal) / (1 - (1 + monthly_rate) ** -months)
  else:
      monthly_payment = principal / months
  return monthly_payment
```

```python
class Point:
  def __init__(self, x, y):
    self.x = x
    self.y = y

  def distance(self, other):
    dx = (self.x - other.x) ** 2
    dy = (self.y - other.y) ** 2
    return (dx + dy) ** 0.5

class Rectangle:
  def __init__(self):
    self.p1 = Point(0,0)
    self.p2 = Point(5,0)
    self.p3 = Point(0,5)
    self.p4 = Point(5,5)
    self.width = self.p1.distance(self.p2)
    self.height = self.p1.distance(self.p3)

  def area(self):
    return self.width * self.height
```
