Assignment #1

```python
sales = float(input("Enter sales ($): "))
tax_rate = float(input("Enter tax rate (%): "))

cogs = 0.4 * sales
gp = sales - cogs
taxes = gp * tax_rate / 100
net = gp - taxes

print(f"Sales: ${sales: }$")
print(f"Cost of Goods Sold: ${cogs: }")
print(f"Gross Profit: ${gp: }$")
print(f"Taxes: {taxes}$")
print(f"Net Income : {net}$")
```
